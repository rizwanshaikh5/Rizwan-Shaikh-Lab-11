```python
class ResidentialHouses(object):
    def __init__(self,location,rooms,parking,price,type):
        self.location=location
        self.rooms =rooms
        self.parking=parking
        self.price=price
        self.type=type

    def single(self):
        print("Type:",self.type)
        print("Location:",self.location)
        print("Rooms:",self.rooms)
        print("Parking:",self.parking)
        print("Price:",self.price)

print("\tSINGLE FAMILY")
class SingleFamily(ResidentialHouses):
    pass
house=SingleFamily("Rabia city",4,"no",1000000,"Single Family")
house.single()
print("\tJOINT FAMILY")
class JointFamily(ResidentialHouses):
    pass
house=JointFamily("Nazimabad",8,"yes",2500000,"Joint Family")
house.single()
print("\tFLATES")
class Flates(ResidentialHouses):
    pass
house=Flates("Eid Ghah",3,"yes",1500000,"Flates")
house.single()
print("\tAPARTMENTS")
class Apartments(ResidentialHouses):
    pass
house=Apartments("Defence",10,"yes",3000000,"Apartments")
house.single() 
```

```python
class ToyotaMotors(object):
    def __init__(self,model,color,price,type,yoManu,yoRegis):
        self.model=model
        self.color=color
        self.price=price
        self.type=type
        self.yoManu=yoManu
        self.yoRegis=yoRegis
print("\tToyota Corolla")
class Corolla(ToyotaMotors):
    def lst(self):
        print("Type:",self.type)
        print("Model:",self.model)
        print("Colour",self.color)
        print("Price:",self.price)
        print("Year Of Manufacture:",self.yoManu)
        print("Year Of Registration:",self.yoRegis)
type1=Corolla(2015,"Black",1600000,"Saloon",2015,2016)
type1.lst()
print("\tToyota Prius")
class Prius(ToyotaMotors):
    def lst(self):
        print("Type:",self.type)
        print("Model:",self.model)
        print("Colour",self.color)
        print("Price:",self.price)
        print("Year Of Manufacture:",self.yoManu)
        print("Year Of Registration:",self.yoRegis)
type1=Prius(2018,"Grey",2500000,"Luxury",2018,2018)
type1.lst()
print("\tToyota Fortuner")
class Fortuner(ToyotaMotors):
    def lst(self):
        print("Type:",self.type)
        print("Model:",self.model)
        print("Colour",self.color)
        print("Price:",self.price)
        print("Year Of Manufacture:",self.yoManu)
        print("Year Of Registration:",self.yoRegis)
type1=Fortuner(2019,"White",4000000,"Luxury",2019,2020)
type1.lst()
print("\tToyota Hilux")
class Hilux(ToyotaMotors):
    def lst(self):
        print("Type:",self.type)
        print("Model:",self.model)
        print("Colour",self.color)
        print("Price:",self.price)
        print("Year Of Manufacture:",self.yoManu)
        print("Year Of Registration:",self.yoRegis)
type1=Hilux(2014,"Black",3500000,"Saloon",2014,2015)
type1.lst()
print("\tToyota Prado")
class Prado(ToyotaMotors):
    def lst(self):
        print("Type:",self.type)
        print("Model:",self.model)
        print("Colour",self.color)
        print("Price:",self.price)
        print("Year Of Manufacture:",self.yoManu)
        print("Year Of Registration:",self.yoRegis)
type1=Prado(2017,"Silver",2500000,"Sport",2017,2018)
type1.lst()
```

```python
class plant(object):
    def __init__(self,name,type,color,imported,taste):
        self.name=name
        self.type=type
        self.color=color
        self.imported=imported
        self.taste=taste
    def final(self):
        print("Name :",self.name)
        print("Type :",self.type)
        print("Color :",self.color)
        print("Imported or Local :",self.imported)
        print("Taste :",self.taste)
print("-------------------Strawberry plant------------")
class Strawberry(plant):
    pass
plant1=Strawberry("Strawberry","Fruit","Red","Local","Sweet")
plant1.final()
print("-------------------Watermelon plant------------")
class Watermelon(plant):
    pass
plant1=Watermelon("Watermelon","Fruit","Red","Local","Sweet")
plant1.final()
print("-------------------Grapes plant------------")
class Grapes(plant):
    pass
plant1=Grapes("Grapes","Fruit","Black","Local","Sweet")
plant1.final()
print("-------------------Rhubarb plant------------")
class Rhubarb(plant):
    pass
plant1=Rhubarb("Rhubarb","Poisonous","Pink","imported","Sour")
plant1.final()
print("-------------------Hydrangea plant------------")
class Hydrangea(plant):
    pass
plant1=Hydrangea("Hydrangea","Poisonous","Purple","imported","Sour")
plant1.final()
```

```python
class Student(object):
    def __init__(self,Student_ID,Student_Name,Department,Section,Address,Number):
        self.Student_ID=Student_ID
        self.Student_Name=Student_Name
        self.Department=Department
        self.Section=Section
        self.Address=Address
        self.Number=Number
    def Final(self):
        print("Student_ID:",self.Student_ID)
        print("Student_Name:",self.Student_Name)
        print("Department:",self.Department)
        print("Section:",self.Section)
        print("Address:",self.Address)
        print("Number:",self.Number)
print("----------------Student Class------------------")
Student1=Student(37,"Rizwan Shaikh","SE","A","Rabia city",3142309877)
Student1.Final()
print("----------------Teacher Class------------------")
class Teacher(object):
    def __init__(self,Teacher_ID,Teacher_Name,Department,Salary,Address,Number):
        self.Teacher_ID=Teacher_ID
        self.Teacher_Name=Teacher_Name
        self.Department=Department
        self.Salary=Salary
        self.Address=Address
        self.Number=Number
    def Final(self):
        print("Teacher_ID:",self.Teacher_ID)
        print("Teacher_Name:",self.Teacher_Name)
        print("Department:",self.Department)
        print("Salary:",self.Salary)
        print("Address:",self.Address)
        print("Number:",self.Number)
Student1=Teacher(1,"Sir Faisal","SE",150000,"Airport",90078601)
Student1.Final()
```
